# Importing Libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# Import dataset

```python
data = pd.read_excel('Spam Email Detection.xlsx')
```

```python
data.head()
```

```python
df = data.where((pd.notnull(data)), '')
```

```python
data = df.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis = 1)
```

```python
data
```

```python
data.shape
```

```python
data.columns
```

# Convert strings to binary

```python
data.loc[data['v1'] == 'spam', 'v1',] = 0
data.loc[data['v1'] == 'ham', 'v1',] = 1
```

```python
X = data['v2']

Y = data['v1']
```

```python
X
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 3)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
print(Y.shape)
print(Y_train.shape)
print(Y_test.shape)
```

# Convert int in X_train into strings

```python
inte = False
for x in X_train:
    if isinstance(x, int):
        inte = True
        break
print(inte)
```

```python
X_train = [str(x) if isinstance(x, int) else x for x in X_train]
```

```python
feature_extraction = TfidfVectorizer(min_df = 1, stop_words = 'english', lowercase = True)

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)

Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
X_train
```

```python
print(X_train_features)
```

# Train the model

```python
model = LogisticRegression()
```

```python
model.fit(X_train_features, Y_train)
```

```python
prediction_on_training_data = model.predict(X_train_features)
accuracy_of_training_data = accuracy_score(Y_train, prediction_on_training_data)
```

```python
print(f"Accuracy of tarining data : {accuracy_of_training_data}")
```

```python
prediction_on_test_data = model.predict(X_test_features)
accuracy_of_test_data = accuracy_score(Y_test, prediction_on_test_data)
```

```python
print(f"Accuracy of test data : {accuracy_of_test_data}")
```

# Building the Predictive System

```python
input_your_mail = ["Free entry in 2 a wkly comp to win FA Cup final tkts 21st May 2005. Text FA to 87121 to receive entry question(std txt rate)T&C's apply 08452810075over18's"]

input_data_features = feature_extraction.transform(input_your_mail)

prediction = model.predict(input_data_features)

print(prediction)

if (prediction == 1):
    print('Ham mail')

else:
    print('Spam Mail')
```
